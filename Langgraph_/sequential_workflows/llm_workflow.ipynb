```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()
```

```python
llm = ChatGroq(model = "llama-3.1-8b-instant")
```

```python
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def llm_response(state: LLMState) -> LLMState:
    question = state["question"]

    prompt = f"Answer to the question: {question}"

    response = llm.invoke(prompt)

    state["answer"] = response.content

    return state
```

```python
graph = StateGraph(LLMState)

graph.add_node("llm_response", llm_response)

graph.add_edge(START, "llm_response")
graph.add_edge("llm_response", END)

workflow = graph.compile()
```

```python
initial_state = {"question": "What is the capital of Ukraine?"}
final_state = workflow.invoke(initial_state)["answer"]
final_state
```
