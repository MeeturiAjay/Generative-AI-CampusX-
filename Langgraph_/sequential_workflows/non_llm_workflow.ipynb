```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# state creation

class BMIState(TypedDict):
    height: float
    weight: float
    bmi: float
    category: str
```

```python
def bmi_calculation(state: BMIState) -> BMIState:
    state["bmi"] = round(state["weight"]/(state["height"]**2), 2)
    return state

def bmi_label(state: BMIState) -> BMIState:
    if state["bmi"] <= 20:
        state["category"] = "not fit"
    else:
        state["category"] = "fit"

    return state
```

```python
# graph object/instance creation
graph = StateGraph(BMIState)

# graph nodes creation
graph.add_node("bmi_calculation", bmi_calculation)
graph.add_node("bmi_label", bmi_label)

# graph edges creation
graph.add_edge(START, "bmi_calculation")
graph.add_edge("bmi_calculation", "bmi_label")
graph.add_edge("bmi_label", END)

# graph compilation
workflow = graph.compile()
```

```python
initial_state = {"weight": 85, "height": 1.82}
final_state = workflow.invoke(initial_state)
print(final_state)
```
