```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()
```

```python
class LLMState(TypedDict):
    topic: str
    outline: str
    blog: str
```

```python
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
```

```python
def generate_outline(state: LLMState) -> LLMState:
    state["outline"] = llm.invoke(f"generate an outline for the topic: {state['topic']}").content
    return state

def generate_blog(state: LLMState) -> LLMState:
    state["blog"] = llm.invoke(f"Generate a blog for the topic {state['topic']} using the outline {state['outline']}").content
    return state
```

```python
graph = StateGraph(LLMState)

graph.add_node("generate_outline", generate_outline)
graph.add_node("generate_blog", generate_blog)

graph.add_edge(START, "generate_outline")
graph.add_edge("generate_outline", "generate_blog")
graph.add_edge("generate_blog", END)

workflow = graph.compile()
```

```python
initial_state = {"topic": "Chandrayan"}
final_state = workflow.invoke(initial_state)["blog"]
final_state
```
