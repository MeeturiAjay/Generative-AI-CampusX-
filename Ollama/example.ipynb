```python
from langgraph.graph import StateGraph, START, END
from langchain_ollama import ChatOllama
from dotenv import load_dotenv
from typing import TypedDict

load_dotenv()
```

```python
class ExampleState(TypedDict):
    name: str
    cricketers: list
```

```python
graph = StateGraph(ExampleState)
llm = ChatOllama(model = "llama3.2:latest")
```

```python
def get_cricketers_with_names(state: ExampleState) -> ExampleState:
    prompt = """
    Give me a list of cricketers with surname Smith.
    Return ONLY a Python list of names.
    Example: ["Name1", "Name2"]
    """
    
    state["cricketers"] = llm.invoke(prompt).content
    return state
```

```python
graph.add_node("Get cricketers with names", get_cricketers_with_names)

graph.add_edge(START, "Get cricketers with names")
graph.add_edge("Get cricketers with names", END)

workflow = graph.compile()
```

```python
initial_state = {"name": "Smith"}
print(workflow.invoke(initial_state)["cricketers"])
```
